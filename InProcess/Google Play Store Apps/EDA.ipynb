```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_Apps = pd.read_csv('Data\googleplaystore.csv')
df_Apps_Rev = pd.read_csv('Data\googleplaystore_user_reviews.csv')
```

```python
df_Apps_Rev
```

```python
df_Apps.info()
```

```python
df_Count_Categories = df_Apps.groupby('Category')['App'].count().reset_index()
df_Count_Categories.index
```

```python
sns.set(rc = {'figure.figsize':(50,20)})
sns.set(style='whitegrid', palette='icefire', font="DejaVu Sans", font_scale=1.3)
sns.barplot(df_Count_Categories[df_Count_Categories['App']>=250], x='Category',y='App')
```

```python
sns.set(rc = {'figure.figsize':(50,20)})
sns.set(style='whitegrid', palette='icefire', font="DejaVu Sans", font_scale=1.3)
sns.barplot(df_Count_Categories[df_Count_Categories['App']<=250], x='Category',y='App')
```
