```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
import pandas as pd
import numpy as np
import plotly.express as px
```

# Creacion y Filtrado

```python
Df_Original = pd.read_csv('Datos/HNP_StatsData.csv')
Df_Original.head(1)
```

```python
Df_Original=Df_Original.drop(['Unnamed: 65'],axis=1)# elimino la columna unnamed
Df_Original.head(1)
```

```python
#Df_IndCode =Df_Original[Df_Original['Indicator Code'].isin(['SP.DYN.CBRT.IN','SP.DYN.LE00.IN','HD.HCI.OVRL','NY.GNP.PCAP.CD'])]
#Df_IndCode=Df_Origin.loc[(Df_Origin['Indicator Code'] == 'SP.DYN.CBRT.IN')|(Df_Origin['Indicator Code'] == 'SP.DYN.LE00.IN')|(Df_Origin['Indicator Code'] == 'HD.HCI.OVRL')|(Df_Origin['Indicator Code'] == 'NY.GNP.PCAP.CD')]
# Me quedo con estos indicadores especificos que mirando en la pagina de donde saque los datos, todos hablan de la taza de nacimiento o cantidad de poblacion.
Df_IndCode=Df_Original[Df_Original['Indicator Code'].str.contains('SP.DYN.CBRT.IN$|SP.DYN.LE00.IN$|HD.HCI.OVRL$|NY.GNP.PCAP.CD$', regex=True)]
Df_IndCode.head(1)
```

```python
#hago un pivot de todas la columnas de años para trasformarlas en filas asi el data set es mas procesable
Df_Trans = Df_IndCode.melt(id_vars=Df_IndCode.columns[:4],value_vars=Df_IndCode.columns[4:],var_name='Año',value_name='Cantidad')
Df_Trans = Df_Trans.round(3)
Df_Trans.dtypes
```

```python
Df_Trans['Año'] = Df_Trans['Año'].astype('int') # transformo el tipo de dato año objeto a un tipo entero
Df_Trans.info()
```

```python
Df_Trans['Intervalo 5'] = pd.cut(Df_Trans['Año'],12) # corto los años en 12 intervalos esto es para que vaya de 5 en 5
Df_Trans['Intervalo 10'] = pd.cut(Df_Trans['Año'],6) # hago lo mismo que arriba solo que el intervalo ahora es de 10 en 10
Df_Trans
```

```python
No_paises1 = ['VGB','UMC','INX','ARB','NAC','CSS','CEB','PRK','CHI','EAR','EAS','EAP','TEA','EUU','TEC','ECA','ECS','EMU','FCS','HCP','HPC','HIC','LAO','LTE','LCN','LAC','TLA','LDC','LMY','LIC','LMC','TMN','MNA','MEA','MIC','OED','OSS','PSS','PRE','PST','SST','SAS','TSA','SSA','SSF','TSS']
#Creo una lista de todo lo que no es un pais
DF_No_paises = Df_Trans['Country Code'].isin(No_paises1)
#elimino todo lo que no seria un pais excepto WLD que seria el total mundial
Df_Trans_Si_Pais = Df_Trans[~DF_No_paises]
Df_Trans_Si_Pais
```

```python
Df_Trans_No_Pais = Df_Trans[DF_No_paises] # creo un DF de lo que no seria un pais si no continentes entre otros.
Df_Trans_No_Pais
```

# Graficos Paises

```python
Limitrofes = ['ARG','URY','CHL','BRA','BOL','USA','CHN','ESP','AUS'] # lista de paises limitrofes a argentina
Df_TasaNacimientos_paises = Df_Trans_Si_Pais[(Df_Trans_Si_Pais['Indicator Code']=='SP.DYN.CBRT.IN') & (Df_Trans_Si_Pais['Country Code'].isin(Limitrofes))]
Df_EsperanzaDeVida_paises = Df_Trans_Si_Pais[(Df_Trans_Si_Pais['Indicator Code']=='SP.DYN.LE00.IN') & (Df_Trans_Si_Pais['Country Code'].isin(Limitrofes))]
Df_PBI_paises = Df_Trans_Si_Pais[(Df_Trans_Si_Pais['Indicator Code']=='NY.GNP.PCAP.CD') & (Df_Trans_Si_Pais['Country Code'].isin(Limitrofes))]
Df_CapitalHumano_paises = Df_Trans_Si_Pais[(Df_Trans_Si_Pais['Indicator Code']=='HD.HCI.OVRL') & (Df_Trans_Si_Pais['Country Code'].isin(Limitrofes))]
```
